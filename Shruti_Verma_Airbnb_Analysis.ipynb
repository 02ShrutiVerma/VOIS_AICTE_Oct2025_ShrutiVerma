```python
# # Airbnb Data Analysis Project

# ## 1. Setup and Data Loading
# Import necessary libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style and default size for all plots
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 7)

# ---
# Load the clean dataset.
try:
    # We are now reading the clean file you created with Google Sheets.
    df = pd.read_csv('/content/airbnb_data.csv.csv')

    print("✅ Dataset loaded successfully!")
    print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
except FileNotFoundError:
    print("❌ Error: The file 'airbnb_data.csv' was not found.")
    print("Please make sure you have uploaded the clean CSV file you created.")
    df = pd.DataFrame()
except Exception as e:
    print(f"An error occurred: {e}")
    df = pd.DataFrame()

# ---
# ## 2. Initial Data Inspection
if not df.empty:
    print("\nFirst 5 rows of the dataset:")
    display(df.head())
    df.info()

# ---
# ## 3. Data Cleaning and Preprocessing
if not df.empty:
    if 'license' in df.columns:
        df.drop('license', axis=1, inplace=True)

    df.dropna(subset=['host name', 'Construction year', 'price', 'service fee'], inplace=True)
    df['reviews per month'].fillna(0, inplace=True)

    df['price'] = df['price'].astype(str).str.replace(r'[\$,]', '', regex=True).astype(float)
    df['service fee'] = df['service fee'].astype(str).str.replace(r'[\$,]', '', regex=True).astype(float)
    df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
    df['Construction year'] = df['Construction year'].astype(int)

    initial_rows = len(df)
    df.drop_duplicates(inplace=True)
    print(f"\nRemoved {initial_rows - len(df)} duplicate rows.")
    print(f"The cleaned dataset now has {df.shape[0]} rows.")

# ---
# ## 4. Exploratory Data Analysis (EDA)
if not df.empty:
    # Q1: Neighborhood Groups
    plt.figure()
    sns.countplot(y=df['neighbourhood group'], order=df['neighbourhood group'].value_counts().index, palette='viridis')
    plt.title('Number of Listings by Neighborhood Group')
    plt.show()

    # Q2: Room Types
    plt.figure()
    room_type_counts = df['room type'].value_counts()
    plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('plasma'))
    plt.title('Distribution of Room Types')
    plt.show()

    # Q3: Average Price by Neighborhood
    plt.figure()
    avg_price_by_neighborhood = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False)
    sns.barplot(x=avg_price_by_neighborhood.index, y=avg_price_by_neighborhood.values, palette='cubehelix')
    plt.title('Average Price by Neighborhood Group')
    plt.xticks(rotation=45)
    plt.show()

    # Q4: Price by Room Type
    plt.figure()
    sns.boxplot(x='room type', y='price', data=df, palette='magma')
    plt.title('Price Distribution by Room Type')
    plt.ylim(0, df['price'].quantile(0.95))
    plt.show()

    # Q5: Correlation Matrix
    cols_to_correlate = [
        'Construction year', 'price', 'service fee', 'minimum nights',
        'number of reviews', 'reviews per month', 'review rate number',
        'calculated host listings count', 'availability 365'
    ]
    existing_cols = [col for col in cols_to_correlate if col in df.columns]
    correlation_matrix = df[existing_cols].corr()
    plt.figure(figsize=(14, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
    plt.title('Correlation Matrix of Numerical Features')
    plt.show()

    # Q6: Price vs. Reviews
    plt.figure()
    sns.scatterplot(x='number of reviews', y='price', data=df, alpha=0.5, hue='room type')
    plt.title('Price vs. Number of Reviews')
    plt.ylim(0, df['price'].quantile(0.98))
    plt.show()

    # Q7: Availability Distribution
    plt.figure()
    sns.histplot(df['availability 365'], bins=50, kde=True, color='teal')
    plt.title('Distribution of Listing Availability (Days per Year)')
    plt.show()

    # Q8: Top 10 Reviewed Listings
    plt.figure(figsize=(12, 8))
    top_10_reviewed = df.nlargest(10, 'number of reviews')
    sns.barplot(x='number of reviews', y='NAME', data=top_10_reviewed, palette='rocket')
    plt.title('Top 10 Most Reviewed Listings')
    plt.show()

    print("\n--- ✅ End of Analysis ---")
```
